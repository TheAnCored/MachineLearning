```python
# Import encoder and pandas
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
# Create simple dataframe
data = {
    'product':['phone', 'tv', 'headphones', 'laptop', 'tablet', 'microsoft'],
    'category':['smartphones', 'license', 'electronics', 'electronics', 'electronics','license'],
    'price':[700,20,300,1000,800,40]
}

df = pd.DataFrame(data)
df
```

```python
# object OHE
encoder = OneHotEncoder(sparse_output=False) # sparse parameter used for the np.array output
```

```python
# transform some features
encoded_product = encoder.fit_transform(df[['product']])

# get output columns
encoded_columns = encoder.get_feature_names_out(['product'])

# setup new dataframe with new columns
encoded_df = pd.DataFrame(encoded_product, columns=encoded_columns)
```

```python
# insert new columns to the df
df = pd.concat([df, encoded_df], axis=1)

# and drop the older 'product' feature from df
df.drop(columns='product', axis=0)
```
